# 手法の説明
このモデルは，低振幅のベッドウェーブが移動しながら累重することで平行葉理が形成されるプロセスをシミュレートするものである．いま，ベッドフォームが速度$c$で移動しつつ速度$D$で累重するとすると，砂床高$\eta$の時空変化は以下のようになる：


\begin{equation}
\frac{\partial \eta}{\partial t} + c \frac{\partial \eta}{\partial x} = D \tag{1}
\end{equation}


ここで，$x$は流向方向の空間座標であり，下流方向を正とする．ただし，ベッドフォームは移動しながら形状を変化させることはないものとし，ベッドフォームの累重速度は空間的に一定であるものとする．式 (1) を積分して解くと，

\begin{equation}
\eta(t, x) = \eta(0, x - ct) + \int_0^t D dt \tag{2}
\end{equation}

となる．すなわち，時刻$t = 0$のときのベッドフォームの形状および累重速度$D(t)$を定義すれば，ある地点$x$の砂床高$\eta$を求めることができる．

## ベッドフォームの定義
本研究では，平行葉理を作るベッドフォームの形状を正弦波で定義する．すなわち，

\begin{equation}
\eta(0,x) = A \sin(\frac{2 \pi}{\lambda} x) \tag{3}
\end{equation}

である．ここで，$A$はベッドフォームの振幅，$\lambda$はベッドフォームの波長である．これを式 (2) に代入すると，

\begin{equation}
\eta(t,x) = A(x - ct) \sin(\frac{2 \pi}{\lambda} (x - ct)) + \int_0^t D dt \tag{4}
\end{equation}


なお，ベッドフォームを構成する堆積物の粒度だが，一般に低振幅ベッドウェーブのクレストでは粒径が粗く，トラフでは細かいことが知られている．そこで，ここでは簡単のため，堆積物の平均粒径は以下のようになることを想定する：

\begin{equation}
d(0, x) = - \alpha_1 \sin(\frac{2 \pi}{\lambda} x) + \alpha_2 \tag{5}
\end{equation}

ここで，$\alpha_1, \alpha_2$は粒径を決める適当な係数である．

## 無次元化
このモデルのパラメーターをベッドフォーム波長$\lambda \mathrm{[L]}$およびベッドフォーム移動速度$c\mathrm{[LT^{-1}]}$で無次元化する．すなわち，

\begin{eqnarray}
\bar{x} = \frac{x}{\lambda},\\
\bar{t} = \frac{c t}{\lambda},\\
\bar{\eta} = \frac{\eta}{\lambda},\\
\bar{D} = \frac{D}{c},\\
\bar{A} = \frac{A}{\lambda}.
\end{eqnarray}

となる．つまり，

\begin{eqnarray}
x = \lambda \bar{x},\\
t = \frac{\lambda \bar{t}} {c},\\
\eta = \lambda \bar{\eta},\\
D = c \bar{D},\\
A = \lambda \bar{A}.
\end{eqnarray}

である．ここで，$\bar{x},\bar{t},\bar{\eta},\bar{D}$はそれぞれ無次元空間座標，無次元時間，無次元砂床高，無次元ベッドフォーム累重速度である．この無次元化によって，式\ref{eq:basic2}は以下のようになる：

\begin{equation}
\bar{\eta}(\bar{t}, \bar{x}) = \bar{A}(\bar{x} - \bar{t})) \sin(2 \pi (\bar{x} - \bar{t})) + \int_0^\bar{t} \bar{D} d\bar{t} \tag{6}
\end{equation}
